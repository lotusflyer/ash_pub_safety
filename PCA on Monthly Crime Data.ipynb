
# PCA on Crime Data


Principal Component Analysis applied to the Asheville Crime data



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle


from sklearn import decomposition
#from sklearn import datasets
```

```python
data = pd.read_pickle('pivot_table_crime_by_month.pkl')
```

```python
X = data[['BURGLARY', 'LARCENY - THEFT']].values
Y = data.index
```

```python
np.random.seed(5)

centers = [[1, 1], [-1, -1], [1, -1]]

fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X = pca.transform(X)

plt.set_cmap('nipy_spectral')

ax.scatter(X[:, 0], X[:, 1], 
           s=100,
           cmap=plt.cm.hot, 
           
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
